```python
#list comprehension and index
def card_ranks(cards):
    "Return a list of the ranks, sorted with higher first."
#     mydict=dict(zip(['A','K','Q','J','T'],[14,13,12,11,10])) 
#     z=lambda x: mydict[x] if x in mydict.keys() else int(x)
#     ranks=[z(x) for x in ranks]
    ranks=['--23456789TJQKA'.index(r) for r,s in cards]
    ranks.sort(reverse=True)
    return ranks
card_ranks(['QC', '5D', 'KS', 'KH'])
```

```python
#GENERATOR FUNCTIONS
def ints(start, end = None):
    i = start
    m=0
    if end is None:
        while i>= start:
            m+= -i*(-1)**i
            yield m
            i = i + 1
    else:
        while i <= end:
            m+= -i*(-1)**i
            yield m
            i = i + 1
    

L=ints(0)
```
